Section 1: Input Data

We define:

One Job Description

Three types of resumes:

Strong Candidate
Average Candidate
Weak Candidate


This helps us test our system properly.

```python
job_description = """
Looking for a Data Scientist with skills in Python, Machine Learning, Deep Learning,
NLP, SQL, and experience with TensorFlow or PyTorch.
"""

strong_resume = """
Experienced Data Scientist skilled in Python, Machine Learning, Deep Learning, NLP,
TensorFlow, PyTorch, and SQL. Worked on multiple AI projects.
"""

average_resume = """
Skilled in Python and basic Machine Learning. Some experience with data analysis
and SQL.
"""

weak_resume = """
Basic knowledge of programming. Familiar with MS Excel and simple data entry tasks.
"""
```

Section 2: Skill Extraction


This step extracts important skills from text. We use a simple keyword-based approach to:

Identify relevant technologies

Avoid hallucination

```python
def extract_skills(text):
    keywords = [
        "python", "machine learning", "deep learning",
        "nlp", "sql", "tensorflow", "pytorch"
    ]

    text = text.lower()

    found_skills = []
    for skill in keywords:
        if skill in text:
            found_skills.append(skill)

    return found_skills
```

Section 3: Matching Logic


Now we compare:

Skills required for the job

Skills present in the resume

This helps identify how well the candidate fits the role.

```python
def match_skills(job_skills, resume_skills):
    matched = list(set(job_skills) & set(resume_skills))
    return matched
```

Section 4: Scoring System


We calculate a score based on:

How many required skills are matched

Formula:

Score = (Matched Skills / Total Job Skills) × 100

```python
def calculate_score(job_skills, matched_skills):
    if len(job_skills) == 0:
        return 0

    score = (len(matched_skills) / len(job_skills)) * 100
    return round(score)
```

Section 5: Explanation Generator


This step explains:

Why the score was given

What skills matched

What skills are missing

This makes the system transparent and trustworthy

```python
def generate_explanation(score, matched, missing):
    explanation = f"""
Final Score: {score}/100

Matched Skills:
{matched}

Missing Skills:
{missing}

The score is calculated based on how many required skills from the job description
are present in the candidate's resume.
"""
    return explanation
```

Section 6: Complete Pipeline


Now we combine all steps into a single function. This represents the full AI pipeline.

```python
def evaluate_resume(resume, job_description):

    job_skills = extract_skills(job_description)
    resume_skills = extract_skills(resume)

    matched = match_skills(job_skills, resume_skills)
    missing = list(set(job_skills) - set(matched))

    score = calculate_score(job_skills, matched)

    explanation = generate_explanation(score, matched, missing)

    return score, explanation
```

Section 7: Testing the System

Strong Candidate

```python
score, explanation = evaluate_resume(strong_resume, job_description)

print("----- Strong Candidate -----")
print(explanation)
```

Average Candidate

```python
score, explanation = evaluate_resume(average_resume, job_description)

print("----- Average Candidate -----")
print(explanation)
```

Weak Candidate

```python
score, explanation = evaluate_resume(weak_resume, job_description)

print("----- Weak Candidate -----")
print(explanation)
```

Analysis

The strong candidate matches most required skills

The average candidate has partial knowledge

The weak candidate lacks required technical skills

This shows the system works correctly.
